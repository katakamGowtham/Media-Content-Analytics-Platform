```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
news_df = news_df.sort_values(by = "date", ascending = True)
```

```python
news_df
```

```python
plt.figure(figsize=(10, 12))
sns.countplot(data=news_df, y="category", order=news_df["category"].value_counts().index)
plt.xlabel("Count")
plt.ylabel("Category")
plt.title("News Category Distribution")
plt.show()
```

```python

# Convert 'date' to datetime format (if not already)
news_df["date"] = pd.to_datetime(news_df["date"])

# Define category mappings
category_mapping = {
    "U.S. NEWS": "News & Politics",
    "WORLD NEWS": "News & Politics",
    "POLITICS": "News & Politics",
    "COMEDY": "Entertainment",
    "ENTERTAINMENT": "Entertainment",
    "ARTS & CULTURE": "Entertainment",
    "PARENTING": "Lifestyle & Family",
    "EDUCATION": "Lifestyle & Family",
    "STYLE & BEAUTY": "Lifestyle & Family",
    "SPORTS": "Health & Sports",
    "HEALTHY LIVING": "Health & Sports",
    "WELLNESS": "Health & Sports",
    "FOOD & DRINK": "Health & Sports",
    "BUSINESS": "Business & Science",
    "MONEY": "Business & Science",
    "SCIENCE": "Business & Science",
    "ENVIRONMENT": "Business & Science",
    "TECHNOLOGY": "Technology",
    "TECH": "Technology"
}

# Apply the category mapping
news_df["grouped_category"] = news_df["category"].map(category_mapping).fillna("Other")

# Count occurrences of each grouped category
category_counts = news_df["grouped_category"].value_counts()

# Plot the grouped count plot
plt.figure(figsize=(12, 6))
sns.barplot(x=category_counts.index, y=category_counts.values, palette="viridis")
plt.xticks(rotation=45, ha="right")
plt.xlabel("Grouped News Category")
plt.ylabel("Count")
plt.title("Grouped News Category Distribution")
plt.show()
```

```python
if "unnamed: 0" in df.columns:
    df.drop("unnamed: 0", axis=1, inplace=True)
```

```python
df.columns = df.columns.str.lower()
df.columns
```

```python
df.to_csv("youtube_data.csv", index = False)
```

```python
# Convert the entire 'Published_date' column to datetime format
df["published_date"] = pd.to_datetime(df["published_date"], errors="coerce").dt.date
```

```python
df["category"].value_counts()
```

```python
plt.figure(figsize = (12,6))
sns.countplot(data = df, y = "category")
```

```python
import matplotlib.ticker as ticker

plt.figure(figsize=(12, 6))
ax = sns.barplot(data=df, y="category", x="views")

# Force normal number formatting (no log scale)
ax.xaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.0f}"))

plt.xlabel("Total Views")
plt.ylabel("Category")
plt.title("Total Views by Category")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
ax = sns.barplot(data=df, y="category", x="likes")

# Force normal number formatting (no log scale)
ax.xaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.0f}"))

plt.xlabel("Total Likes")
plt.ylabel("Category")
plt.title("Total Likes by Category")
plt.show()
```
